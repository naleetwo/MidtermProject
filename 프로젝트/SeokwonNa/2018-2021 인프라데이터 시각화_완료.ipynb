### 자치구별 변화율 요인별 Heatmap 으로 시각화

```python
df.head()
```

```python
#아파트 매매가변동률(기준 = 2018년 데이터) 내림차순으로 정렬

df1 = df.sort_values(by='매매가변동률', ascending=False)
df1
```

```python
#아파트 전월세가변동률(기준 = 2018년 데이터) 내림차순으로 정렬

df2 = df.sort_values(by = '전월세가변동률', ascending = False)
df2.round(2)
#df2.to_csv(r"C:/Users/acorn/OneDrive/Desktop/중간프로젝트/인프라_데이터.csv", encoding= 'utf-8')
```

```python
#자치구별 heatmap, 전월세가변동율 내림차순 기준

fig = plt.subplots(figsize=(8, 9))
cmap = sns.color_palette("rocket", as_cmap=True )
sns.heatmap(data=df2, cmap=cmap, annot=False, square=False)
```

```python
#자치구별 heatmap, 매매가변동율 내림차순 기준

fig = plt.subplots(figsize=(8, 9))
cmap = sns.color_palette("crest", as_cmap=True)
sns.heatmap(data=df1, cmap=cmap, annot=False, square=False, vmax = 80)
```
