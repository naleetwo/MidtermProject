```python
#아파트 매매 실거래가 가져오기 및 필요없는 행 제거

df = pd.read_csv(
    r"C:/Users/acorn/OneDrive/Desktop/중간프로젝트/부동산/아파트(매매)__실거래가_2022_국토교통부.csv", encoding='euc-kr',
        skiprows= 15)
df.head()
```

```python
#데이터 정보 확인
df.info()
```

```python
df1 = df.copy()
```

```python
df1.head()
```

```python
#계약년월 연도.월로 변환
df1['계약년월'] = df1['계약년월'].astype(
    str).apply(lambda x: x[:4] + '.' + x[4:6])
df1.info()
df1['계약년월']
```

```python
df1.columns
```

```python
# 거래금액(만원) column을 숫자형으로 변환
df1['거래금액(만원)'] = df1['거래금액(만원)'].str.replace(',', '').astype(int)
```

```python
#결측치 확인
df1['거래금액(만원)'].isnull().sum()
```

```python
df1['거래금액(만원)'].describe().round(0)
```

```python
#결측데이터 확인
sorted_df1 = df1.sort_values(by='거래금액(만원)', ascending=True)
sorted_df1
```

```python
# 컬럼의 순서(위치) 바꾸기
cols = ['시군구', '전용면적(㎡)', '계약년월','거래금액(만원)']
df2 = pd.DataFrame(df1, columns = cols)
df2
```

```python
df2.head()
```

```python
# 전용면적당 거래금액(만원)
df2['1m^2 전용면적당 거래금액(만원)'] = (df2['거래금액(만원)'] / df2['전용면적(㎡)']).astype(int)
df2.head()
```

```python
#시군구 column 자치구별로 변동
df2.rename(columns = {'시군구':'자치구별'}, inplace = True)
```

```python
df2.head()
```

```python
#자치구별 주소 구로만 가져오기
df2['자치구별'] = df2['자치구별'].str.split().str[1]
df2.head()
```

```python
#계약년월 -> 계약연도, 계약월로 나누기
df2['계약년월'] = df2['계약년월'].str.replace(".", "").astype(int)
df2['계약연도'] = df2['계약년월'] // 100
df2['계약월'] = df2['계약년월'] % 100
```

```python
df2
```

```python
#데이터 자료 분기별로 분류 - 1,2,3,4분기 계약월 기준으로 분류
#계약월 1,2,3 --> Q1 / 4,5,6 --> Q2 / 7,8,9 --> Q3 /10, 11, 12 --> Q4
Q1 = [1, 2, 3]
Q2 = [4, 5, 6]
Q3 = [7, 8, 9]
Q4 = [10, 11, 12]

# 계약월을 분기로 변환하는 함수 정의

def get_quarter(month):
    if month in Q1:
        return 'Q1'
    elif month in Q2:
        return 'Q2'
    elif month in Q3:
        return 'Q3'
    else:
        return 'Q4'

# 분기 컬럼 생성
df2['분기'] = df2['계약월'].apply(get_quarter)

df2
```

```python
df2.columns
```

```python
# 컬럼의 순서(위치) 바꾸기
cols = ['자치구별', '계약연도', '계약월', '분기', '1m^2 전용면적당 거래금액(만원)']
df3 = pd.DataFrame(df2, columns=cols)
df3
```

```python
df3.columns
cols = ['자치구별', '계약연도', '분기', '1m^2 전용면적당 거래금액(만원)']
df3 = pd.DataFrame(df3, columns=cols)
df3
```

```python
# 연도/분기로 계정학목 row 생성
df3['계정항목'] = df3.apply(lambda row: f"{str(row['계약연도'])}/{str(row['분기'])}", axis=1)
df3
```

```python
df3['계정항목'].unique()
```

```python

cols = ['계정항목','자치구별', '1m^2 전용면적당 거래금액(만원)']
df4 = pd.DataFrame(df3, columns=cols)
df4
```

```python
#group_by 사용해서 자치구별 + 계약년월 대비 전용면적당 거래금액(만원) 평균 df으로 구성
df4 = df4.groupby([df4.계정항목]).mean().round(0)
df4
```
