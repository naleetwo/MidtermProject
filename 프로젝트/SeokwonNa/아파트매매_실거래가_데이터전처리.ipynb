```python
#아파트 매매 실거래가 가져오기 및 필요없는 행 제거

df = pd.read_csv(
    r"C:/Users/acorn/OneDrive/Desktop/중간프로젝트/부동산/아파트(매매)__실거래가_2022_국토교통부.csv", encoding='euc-kr',
        skiprows= 15)
df.head()
```

```python
#데이터 정보 확인
df.info()
```

```python
df1 = df.copy()
```

```python
df1.head()
```

```python
df1.columns
```

```python
# 컬럼의 순서(위치) 바꾸기
cols = ['시군구', '번지', '본번', '부번', '단지명', '전용면적(㎡)', '계약년월','거래금액(만원)']
df2 = pd.DataFrame(df1, columns = cols)
df2
```

```python
#거래금액(만원) column str을 float으로 수정
df2['거래금액(만원)'] = df2['거래금액(만원)'].str.replace(',', '').astype(int)
df2['계약년월'] = df2['계약년월'].astype(int)
df2.info()
```

```python
df2.head()
```

```python
# 전용면적당 거래금액(만원)
df2['1m^2 전용면적당 거래금액(만원)'] = (df2['거래금액(만원)'] / df2['전용면적(㎡)']).astype(int)
df2.head()
```

```python
#시군구 column 자치구별로 변동
df2.rename(columns = {'시군구':'자치구별'}, inplace = True)
```

```python
df2.head()
```

```python
#자치구별 주소 구로만 가져오기
df2['자치구별'] = df2['자치구별'].str.split().str[1]
df2.head()
```

```python
#계약년월 -> 계약연도, 계약월로 나누기
df2['계약연도'] = df2['계약년월'] // 100
df2['계약월'] = df2['계약년월'] % 100
```

```python
df2
```

```python
#중간포인트 저장
df2.to_csv(r"C:/seokwonna/프로젝트/save_point1_apt_buysell_2022_data.csv", encoding='utf-8')
```

```python
df2
```

```python
#group_by 사용해서 자치구별 + 계약년월 대비 전용면적당 거래금액(만원) 평균 df으로 구성
df3 = df2.groupby([df2.자치구별, df2.계약월]).mean().round(0)
df3
```

```python
unique_values = df2['자치구별'].unique()
print(unique_values)
print(unique_values.size)
```

```python
#자치구 월별 매매된 평균 아파트 매매가, 1m^2 전용면적당 평균가
df3.to_csv(
    r"C:/seokwonna/프로젝트/apt_buysell_price_2018-22/save_point2_apt_buysell_2022_data.csv", encoding='utf-8')
```
