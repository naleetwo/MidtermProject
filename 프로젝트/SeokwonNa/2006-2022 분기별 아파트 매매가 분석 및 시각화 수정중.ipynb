```python
import plotly.graph_objs as go
import pandas as pd

# 데이터 프레임 df1이 이미 정의되어 있다고 가정합니다.

# 그래프에 사용할 x축을 설정합니다.
x_axis = df1['key_0']

# 각 feature에 해당하는 y축 데이터를 설정합니다.
y1 = df1['1m^2 전용면적당 거래금액(만원)']
y2 = df1['물가지수']
y3 = df1['GDP(십억원)']
y4 = df1['기준금리(연%)']

# 그래프 객체를 생성합니다.
fig = go.Figure()

# trace를 추가합니다.
fig.add_trace(go.Scatter(x=x_axis, y=y1, name='1m^2 전용면적당 거래금액(만원)'))
fig.add_trace(go.Scatter(x=x_axis, y=y2, name='물가지수'))
fig.add_trace(go.Scatter(x=x_axis, y=y3, name='GDP(십억원)'))
fig.add_trace(go.Scatter(x=x_axis, y=y4, name='기준금리(연%)'))

# 그래프 레이아웃을 설정합니다.
fig.update_layout(title='시계열 그래프', xaxis_title='key_0', yaxis_title='값')

# 그래프를 출력합니다.
fig.show()
```
