```python
data1=pd.read_csv("C:\\Users\\owner\\team project\\중학교(국·공립)_20230417163307.csv")
```

```python
del data1["자치구별(1)"]
```

```python
# 컬럼명 변경
data1.columns=["자치구별","2018학교수","2019학교수","2020학교수","2021학교수","2022학교수"]
```

```python
data1
```

```python
# 자치구별 인덱스 설정
data1.set_index("자치구별",inplace=True)
```

```python
# 필요없는 데이터 삭제
data1.drop("자치구별(2)",inplace=True)
data1.drop("소계",inplace=True)
data1.drop("국립",inplace=True)
data1.drop("공립",inplace=True)
```

```python
data1
```

```python
data2=pd.read_csv("C:\\Users\\owner\\team project\\중학교(사립)_20230417163450.csv")
```

```python
data2
```

```python
del data2["자치구별(1)"]
```

```python
# 컬럼명 변경
data2.columns=["자치구별","2018학교수","2019학교수","2020학교수","2021학교수","2022학교수"]
```

```python
# 자치구별 인덱스 설정
data2.set_index("자치구별",inplace=True)
```

```python
data2
```

```python
# 필요없는 데이터 삭제
data2.drop("자치구별(2)",inplace=True)
data2.drop("소계",inplace=True)

```

```python
data1=data1.astype(int)
data2=data2.replace("-",0).astype(int)
```

```python
data1.info()
```

```python
data2.info()
```

```python
middle_school=data1.merge(data2, on="자치구별")
```

```python
# 국공립학교 데이터와 사립학교 데이터 합치기
middle_school=data1+data2
```
