```python
import pandas as pd
import re
```

```python
high=pd.read_csv("C:\\Users\\owner\\Desktop\\데이터 분석 중간 프로젝트\\인프라\\서울시 고등학교 기본정보.csv",encoding="EUC-KR")
```

```python
high.head()
```

```python
high1=high["도로명주소"]
```

```python
print(high1)
```

```python
# 주소에서 구만 추출
list=[]
for i in range(321):
    address = high1[i]
    pattern_gu = re.compile(r"(\S+?[구군])")

    result = re.search(pattern_gu, address)
    print(result.group())
    list.append(result.group())
   
```

```python
high["자치구별"]=list # 자치구별 컬럼 추가
```

```python
high.head()
```

```python
high2=high[["학교명","자치구별"]]
```

```python
high2
```

```python
high_school=high2.groupby("자치구별").count() #자치구별 고등학교 갯수 
```

```python
high_school
```

```python
high_school.columns=["고등학교수"]
```

```python
high_school["고등학교수"]=high_school["고등학교수"].astype(int) # 데이터 정수형으로 변경
```
