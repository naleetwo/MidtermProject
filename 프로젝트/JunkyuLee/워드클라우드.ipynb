```python
news = pd.read_csv("C:/ljk9707/PythonWork/Analysis/data/.csv")
news_title = " ".join(news["제목"])

ko_text = news_title
okt = Okt()

token_ko = okt.nouns(ko_text)
stop_words = ['앞', '끌', '속', '그날', '명', '짓', '구', '불', '채', '중', '못', '량', '치', '더', '것', '주', '선', '위', '안', '요', '반', '전', '내', '률', '위', '변', '후', '사', '옆', '특', '삽', '보', '닉', '머', '맛', '배', '자', '처', '남', '친', '거', '씨', '힐', '롭', "제", "권", "급", '입', '당', '실', '건', '최', '율', "등"]

nouns = []
for word in token_ko:
    if word not in stop_words:
        nouns.append(word)
nouns

token_ko = nouns

ko = nltk.Text(token_ko, name="부동산 관련 기사")
data = ko.vocab().most_common()
print(data)

wc= WordCloud(max_font_size=40, background_color="#ffffff",
              max_words=100, font_path = 'C:/Windows/Fonts/malgun.ttf').generate_from_frequencies(dict(data))

plt.figure(figsize=(12, 12))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
```
