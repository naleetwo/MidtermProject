```python
driver = webdriver.Chrome("chromedriver")
driver.get("https://vod.kbs.co.kr/drama.html")
click = driver.find_element(By.XPATH, "//*[@id='broadcasted-tab']/a").click()
drama = driver.find_elements(By.ID, "broadcasted-list")
box = drama[0].find_elements(By.CLASS_NAME, "txt-box")
kbs_name = []
for i in range(len(box)):
    title = box[i].find_element(By.CLASS_NAME, "tit")
    kbs_name.append(title.text)
    print(title.text)
    time.sleep(2)
    if title.text == "조선미인별전":
        break

kbs_name_ex =[x for x in kbs_name if "KBS" not in x and "연기대상" not in x and "TV소설" not in x]

driver.get("https://www.naver.com")

for names in kbs_name_ex:
    try:
        element = driver.find_element(By.ID, 'query')
        element.send_keys(names)
        element.send_keys(Keys.ENTER)

        search = driver.find_element(By.ID, "main_pack")
        box = search.find_element(By.CLASS_NAME, "cm_info_box")
        info = box.find_elements(By.CLASS_NAME, "txt_3")
        day = info[0].find_element(By.CLASS_NAME, "text")
        watch = info[0].find_element(By.CLASS_NAME, "value")
        tex = box.find_elements(By.CLASS_NAME, "_text")
        
        print(names, day.text, watch.text)
        time.sleep(1)
        driver.back()
    
    except Exception as e:
        print(names, e)
        driver.back()
        continue

driver.close()
```

### KBS 등장인물 뽑아내기

```python
driver = webdriver.Chrome("chromedriver")
driver.get("https://vod.kbs.co.kr/drama.html")
click = driver.find_element(By.XPATH, "//*[@id='broadcasted-tab']/a").click()
drama = driver.find_elements(By.ID, "broadcasted-list")
box = drama[0].find_elements(By.CLASS_NAME, "txt-box")
kbs_name = []
for i in range(len(box)):
    title = box[i].find_element(By.CLASS_NAME, "tit")
    kbs_name.append(title.text)
    print(title.text)
    time.sleep(2)
    if title.text == "조선미인별전":
        break

kbs_name_ex =[x for x in kbs_name if "KBS" not in x and "연기대상" not in x and "TV소설" not in x]

driver.get("https://www.naver.com")
actor_name =[]
for i in kbs_name_ex:
    
    element =driver.find_element(By.ID, 'query')
    element.send_keys("{} 등장인물".format(i))
    element.send_keys(Keys.ENTER)

    box = driver.find_elements(By.CLASS_NAME, "title_box")
    for i in box:
        route = i.find_element(By.CLASS_NAME, "sub_text")
        actor_name.append(route.text)
        time.sleep(2)
    driver.back()
print(actor_name)
```
