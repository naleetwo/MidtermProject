```python
import matplotlib as mlp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#matplot 한글폰트 설정
import matplotlib.font_manager as fm
font_name = fm.FontProperties(fname="C:\Windows\Fonts\GULIM.TTC").get_name()
plt.rc("font", family=font_name)

mlp.rcParams["axes.unicode_minus"] = False
```

```python
# 경제적 요인 전처리 된 데이터 불러오기

df1 = pd.read_csv(
    r"C:/Users/acorn/OneDrive/Desktop/중간프로젝트/부동산/17년치 분기별 요인(원계열).csv", encoding='utf-8', index_col=0)
df1
```

```python
# 경제적 요인 데이터 정규화

import sklearn
from sklearn import preprocessing

min_max_scale = sklearn.preprocessing.MinMaxScaler()  # StandardScaler 표준화된 방식으로 수행

cols = ['면적당거래가격', '물가지수', '기준금리', '실업율', '고용률', 'GDP', '유동성비율', '인구이동률']

X = df1[cols].values
X

X_scaled = min_max_scale.fit_transform(X)
X_scaled  # 정규화 된 결과'

seoul_normal = pd.DataFrame(X_scaled, columns=cols, index=df1.index)
seoul_normal
```

## 전처리된 데이터 시각화

```python
import plotly.graph_objs as go
import pandas as pd

# 데이터 프레임 df1이 이미 정의되어 있다고 가정합니다.

# 그래프에 사용할 x축을 설정합니다.
x_axis = seoul_normal.index

# 각 feature에 해당하는 y축 데이터를 설정합니다.
y1 = seoul_normal['면적당거래가격']
y2 = seoul_normal['물가지수']
y3 = seoul_normal['기준금리']
y4 = seoul_normal['실업율']
y5 = seoul_normal['고용률']
y6 = seoul_normal['GDP']
y7 = seoul_normal['유동성비율']
y8 = seoul_normal['인구이동률']

# 그래프 객체를 생성합니다.
fig = go.Figure()

# trace를 추가합니다.
fig.add_trace(go.Scatter(x=x_axis, y=y1, name='면적당거래가격'))
fig.add_trace(go.Scatter(x=x_axis, y=y2, name='물가지수'))
fig.add_trace(go.Scatter(x=x_axis, y=y3, name='기준금리'))
fig.add_trace(go.Scatter(x=x_axis, y=y4, name='실업율'))
fig.add_trace(go.Scatter(x=x_axis, y=y5, name='고용률'))
fig.add_trace(go.Scatter(x=x_axis, y=y6, name='GDP'))
fig.add_trace(go.Scatter(x=x_axis, y=y7, name='유동성비율'))
fig.add_trace(go.Scatter(x=x_axis, y=y8, name='인구이동률'))



# 그래프 레이아웃을 설정합니다.
fig.update_layout(
    title='시계열 그래프',
    xaxis_title='2006-2022년/분기별',
    yaxis_title='정규화 값',
    width=1500,
    height=700,
    font=dict(family='Arial', size=18),
    xaxis=dict(tickmode='linear',dtick=2)
)


# 그래프를 출력합니다.
fig.show()
```
