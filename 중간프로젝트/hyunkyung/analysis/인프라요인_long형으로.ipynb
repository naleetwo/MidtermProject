### 변동률

```python
df = pd.read_csv('C:/acorn/midproject/analysis/인프라모든요인.csv', index_col=0)
df
```

```python
df1 = df[df['계약연도']==2018]
df2 = df[df['계약연도']==2021]
df2
```

```python
df=pd.merge(df1, df2, on='자치구별')
df
```

```python
df['범죄건수변동률'] = ((df['범죄건수_y'] - df['범죄건수_x']) / df['범죄건수_x']) * 100
df['정류장수변동률'] = ((df['정류장수_y'] - df['정류장수_x']) / df['정류장수_x']) * 100
df['편의점수변동률'] = ((df['편의점수_y'] - df['편의점수_x']) / df['편의점수_x']) * 100
df['카페수변동률'] = ((df['카페수_y'] - df['카페수_x']) / df['카페수_x']) * 100
df['학교수변동률'] = ((df['학교수_y'] - df['학교수_x']) / df['학교수_x']) * 100
df['병원수변동률'] = ((df['병원수_y'] - df['병원수_x']) / df['병원수_x']) * 100
df['학원수변동률'] = ((df['학원수_y'] - df['학원수_x']) / df['학원수_x']) * 100
df['대학교수변동률'] = ((df['대학교수_y'] - df['대학교수_x']) / df['대학교수_x']) * 100
df['매매가변동률'] = ((df['매매가_y'] - df['매매가_x']) / df['매매가_x']) * 100
df['전월세가변동률'] = ((df['전월세가_y'] - df['전월세가_x']) / df['전월세가_x']) * 100
df
```

```python
final = df[['자치구별', '범죄건수변동률', '정류장수변동률', '편의점수변동률', '카페수변동률', '학교수변동률', '병원수변동률', '학원수변동률', '대학교수변동률', '매매가변동률', '전월세가변동률']]
```
