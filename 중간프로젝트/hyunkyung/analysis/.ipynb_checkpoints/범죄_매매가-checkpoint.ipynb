## 범죄&매매가

```python
from functools import reduce
import pandas as pd
dfs = [crime, buy2018, buy2019, buy2020, buy2021]
df = reduce(lambda left, right: pd.merge(left, right, on='자치구별'), dfs)
df
```

```python
df.to_csv('crime_price_18-22.csv')
```

### 상관분석

```python
print(scipy.stats.spearmanr(df['2018'], df['2018가격'])) # p-value 어떻게 보는 것?
print(scipy.stats.spearmanr(df['2019'], df['2019가격']))
print(scipy.stats.spearmanr(df['2020'], df['2020가격']))
print(scipy.stats.spearmanr(df['2021'], df['2021가격']))
```

```python
corr = df.corr(method='spearman')
corr
```

### 회귀분석

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(df['2018'].values.reshape(-1, 1), df['2018가격'].values.reshape(-1, 1))
lr.coef_[0]
```

```python
import statsmodels.api as sm
results = sm.OLS(df['2018가격'], sm.add_constant(df['2018'])).fit()
results.summary()
```

- F-statistic의 p-value 값은 Prob(F-statistic)으로 표현되는데, 이는 0.05보다 작기에 이 회귀식은 회귀분석 모델 전체에 대해 통계적으로 의미가 있다고 볼 수 있습니다.

- 중간쯤에 보면 coef와 변수 x1의 p-value 값이 나와있습니다. 여기서 x1은 bmi이고 이 변수의 p-value가 0.000으로 표기 되어 있기에 0.05보다 작으므로 target을 설명하는데 유의하다고 판단할 수 있습니다.

```python
results = sm.OLS(df['2019가격'], sm.add_constant(df['2019'])).fit()
results.summary()
```

```python
results = sm.OLS(df['2020가격'], sm.add_constant(df['2020'])).fit()
results.summary()
```

```python
results = sm.OLS(df['2021가격'], sm.add_constant(df['2021'])).fit()
results.summary()
```
