```python
import glob
import re
```

```python
# 아파트 구별 연평균 실거래 매매가 2018-2022년 data merge
# 파일 경로가 저장된 폴더 경로
apt_buyprice_path = r"C:/seokwonna/프로젝트/apt_buysell_price_2018-22"
apt_rentprice_path = r"C:/seokwonna/프로젝트/apt_rent_price_2018-22"
# point3가 포함된 csv 파일의 정규표현식
pattern = re.compile('.*point3.*')

# 파일 이름에 pattern이 포함된 모든 CSV 파일의 경로를 검색
all_files = glob.glob(apt_buyprice_path + "/*.csv")
all_files = [file for file in all_files if pattern.match(file)]

# 모든 CSV 파일의 데이터를 담을 빈 데이터프레임 생성
df_buyprice = pd.DataFrame()

# 모든 CSV 파일을 반복하며 데이터를 결합
for file in all_files:
    temp_df = pd.read_csv(file)  # CSV 파일 읽어오기
    df_buyprice = pd.concat([df_buyprice, temp_df])  # 데이터 결합

# 결과 출력
print(df_buyprice)

```

```python
df_buyprice.describe().round(2)
```

```python
#2018-2022년 아파트 매매 실거래가 전처리데이터
df_buyprice.to_csv(r"C:/seokwonna/프로젝트/apt_buysell_price_2018-22/apt_buysell_2018-2022_preprocessingdata.csv", encoding='utf-8')
```
