```python
import numpy as np
import pandas as pd
```

```python
df_2018 = pd.read_csv('C:/acorn/midproject/bus_station_data/preprocessed_data/2018_bus_count.csv', index_col=False)
df_2019 = pd.read_csv('C:/acorn/midproject/bus_station_data/preprocessed_data/2019_bus_count.csv', index_col=False)
df_2020 = pd.read_csv('C:/acorn/midproject/bus_station_data/preprocessed_data/2020_bus_count.csv', index_col=False)
df_2021 = pd.read_csv('C:/acorn/midproject/bus_station_data/preprocessed_data/2021_bus_count.csv', index_col=False)
df_2022 = pd.read_csv('C:/acorn/midproject/bus_station_data/preprocessed_data/2022_bus_count.csv', index_col=False)
```

```python
df_2018.rename(columns={'정류장수':'2018'}, inplace=True)
df_2019.rename(columns={'정류장수':'2019'}, inplace=True)
df_2020.rename(columns={'정류장수':'2020'}, inplace=True)
df_2021.rename(columns={'정류장수':'2021'}, inplace=True)
df_2022.rename(columns={'정류장수':'2022'}, inplace=True)
```

```python
df_2019
```

```python
from functools import reduce

dfs = [df_2018, df_2019, df_2020, df_2021, df_2022]
df_merge = reduce(lambda left, right: pd.merge(left, right, on='gu'), dfs)
```

```python
df_merge
```

```python
df_merge.rename(columns={'gu':'자치구별'}, inplace=True)
```

```python
df_merge.to_csv('C:/acorn/midproject/bus_station_data/final_bus_gu_year.csv')
```
