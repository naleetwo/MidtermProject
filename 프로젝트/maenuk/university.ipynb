```python
list=driver.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody').text.split("\n")
```

```python
list
```

```python
import re
```

```python
location=[]
for i in list:
    pattern1="[가-힇]+구"  
    result=re.findall(pattern1,i)
    if len(result)== 2 : 
        location.append(result[0])
        location.append(result[1])
    else:
        location.append(result[0])
```

```python
len(location)
```

```python
location   
```

```python
uni=pd.DataFrame({"자치구별" : location})
```

```python
uni
```

```python
uni=uni.groupby("자치구별")[["자치구별"]].count()
```
