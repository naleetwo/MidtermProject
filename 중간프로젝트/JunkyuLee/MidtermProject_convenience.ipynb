```python
import numpy as np
import pandas as pd
```

```python
Seoul_CS2 = pd.read_csv("../../Data/JunkyuLee/Seoul_col.csv", index_col=0)
Seoul_CS2
```

```python
Seoul_CS2 = Seoul_CS2.sort_values("인허가일자")
```

```python
Seoul_CS2["업태구분명"].unique()
```

```python
#편의점만 추출
Seoul_CS3 = Seoul_CS2[Seoul_CS2["업태구분명"] == "편의점"]
Seoul_CS3
```

```python
#서울시 구 리스트 생성
gu_list = []
for i in Seoul_CS3["지번주소"].str.split().str[1].unique()[:-1]:
    gu_list.append(i)
print(gu_list[:-1], end=" ")
```

```python
# 월별 마지막날 리스트 생성
import calendar

Date = []
for year in range(2017, 2023):  # 2017년부터 2022년까지 반복
    for month in range(1, 13):  # 1월부터 12월까지 반복
        last_day = calendar.monthrange(year, month)[1]  # 각 달의 마지막 날짜
        Date.append(f"{year}-{month}-{last_day}")
print(Date, end=" ")
```

```python
#월별 마지막 날 기준으로 구별 커피숍 개수 파악 후 데이터 프레임 생성
df = pd.DataFrame(index=gu_list[:-1], columns=Date)
for i, gu in enumerate(gu_list[:-1]):
    print(gu)
    for j, D in enumerate(Date):
        gu_date = Seoul_CS3[(Seoul_CS3["지번주소"].astype(str).str.split(" ").str[1]==gu) & (Seoul_CS3["인허가일자"] <= D) & ((Seoul_CS3["폐업일자"] >= D) | Seoul_CS3["폐업일자"].isnull())]
        count = len(gu_date)
        df.iloc[i, j] = count
        print(count)
```

```python
# 시각화를 위해 데이터를 숫자형으로 변경
df = df.astype("int64")
df.info()
```

```python
# 다른 데이터와의 merge를 위해 index name 설정
df.index.name="자치구별"
```

```python
#생성된 데이터 프레임 파일로 저장
df.to_csv("../../Data/JunkyuLee/gu_month_convenience.csv")
df.to_csv("../../Data/JunkyuLee/gu_month_convenience_euc-kr.csv", encoding="EUC-KR")
```

```python
#매해 마지막 날 리스트 생성
import calendar

Date = []
for year in range(2017, 2023):  # 2017년부터 2022년까지 반복
    last_day = calendar.monthrange(year, 12)[1]  # 각 해의 마지막 날짜
    Date.append(f"{year}-{month}-{last_day}")
print(Date, end=" ")
```

```python
#매해 마지막 날 기준으로 구별 커피숍 개수 파악 후 데이터 프레임 생성 
df_year = pd.DataFrame(index=gu_list[:-1], columns=Date)
for i, gu in enumerate(gu_list[:-1]):
    print(gu)
    for j, D in enumerate(Date):
        gu_date = Seoul_CS3[(Seoul_CS3["지번주소"].astype(str).str.split(" ").str[1]==gu) & (Seoul_CS3["인허가일자"] <= D) & ((Seoul_CS3["폐업일자"] >= D) | Seoul_CS3["폐업일자"].isnull())]
        count = len(gu_date)
        df_year.iloc[i, j] = count
        print(count)
```

```python
# column 변경을 위해 년도만 추출하여 리스트 생성
year = []
for i in pd.Series(df_year.columns).str.split("-"):
    year.append(i[0])
print(year)
```

```python
# column을 년도로 변경
df_year.columns=year
df_year
```

```python
# 시각화를 위해 모든 데이터를 숫자 데이터로 변경
df_year = df_year[:].astype("int64")
```
